```python
import matplotlib.pyplot as plt
import numpy as np

# Dados dos resultados
data = {
    "PSO VGG16": [0.9756082851, 0.9760516026, 0.976612035, 0.9778650022, 0.9736031269, 0.9758420142],
    "PSO Xception": [0.9433017749, 0.9533494316, 0.9465543291, 0.950047628, 0.955402628, 0.9360720388],
    "Grid Search VGG16": [0.9770894177, 0.9801659288, 0.982474924],
    "Random Search VGG16": [0.9763715279, 0.9647729648, 0.9760248878, 0.9769568535]
}

# Criar um grfico de barras para mostrar a sensibilidade dos parmetros
labels = list(data.keys())
means = [np.mean(scores) for scores in data.values()]

plt.figure(figsize=(10, 6))
plt.bar(labels, means, color=['blue', 'green', 'red', 'purple'])
plt.xlabel('Mtodo de Otimizao e Modelo')
plt.ylabel('AUC Mdia')
plt.title('Sensibilidade dos Parmetros nos Diferentes Mtodos de Otimizao')
plt.xticks(rotation=45)
plt.show()
```
